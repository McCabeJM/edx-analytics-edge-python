```python
songTrain = songs[songs['year'] <= 2009]
songTest = songs[songs['year'] > 2009]
print songTrain.shape
print songTest.shape
```

```python
# Create the logistic regression model
# First, specify which variables to use as independent variables
to_exclude = ["year", "songtitle", "artistname", "songID", "artistID"]
ind_vars = [col for col in songs.columns if col not in to_exclude and col != "Top10"]
model_formula = "Top10 ~ " + " + ".join(ind_vars)
SongsLog1 = glm(formula=model_formula, data=songTrain, 
                   family = sm.families.Binomial(sm.families.links.logit)).fit()
print SongsLog1.summary()
print "Akaike Information Criterion (AIC): ", SongsLog1.aic
```

```python
# Find the correlation between loudness and energy in the training set
# We suspect that there might be multicollinearity between these two variables
songTrain.corr().loudness.energy
```

```python
# Create a new model that omits the independent variable loudness
loudness_exclude = to_exclude + ['loudness']
ind_vars2 = [col for col in songs.columns if col not in loudness_exclude and col != "Top10"]
model2_formula = "Top10 ~ " + " + ".join(ind_vars2)
SongsLog2 = glm(formula=model2_formula, data=songTrain, 
                   family = sm.families.Binomial(sm.families.links.logit)).fit()
print SongsLog2.summary()
print "Akaike Information Criterion (AIC): ", SongsLog2.aic
```

```python
# Create a new model that omits the independent variable energy
energy_exclude = to_exclude + ['energy']
ind_vars3 = [col for col in songs.columns if col not in energy_exclude and col != "Top10"]
model3_formula = "Top10 ~ " + " + ".join(ind_vars3)
SongsLog3 = glm(formula=model3_formula, data=songTrain, 
                   family = sm.families.Binomial(sm.families.links.logit)).fit()
print SongsLog3.summary()
print "Akaike Information Criterion (AIC): ", SongsLog3.aic
```

```python
# Same model as above with different method
SongsLog4 = logit(formula=model3_formula, data=songTrain).fit()
SongsLog4.summary()
```

```python
# Make predictions using model 3
predict_test = SongsLog3.predict(songTest)
# Create a confusion matrix
pd.crosstab(songTest.Top10, predict_test >= 0.45)
```

```python
# Function to calculate Specificity and Sensitivity of the model
# as well as the confusion matrix.
# From this we can calculate the accuracy of the model
# To use this method, you need to use the logit method when making
# your model. GLM won't have a pred_table method.
def conf_info(prediction_df, t=0.5):
    conf = prediction_df.pred_table(t)
    print(conf)
    print('Specificity  {0}'.format(conf[0,0]/sum(conf[0])))
    print('Sensitivity {0}'.format(conf[1,1]/sum(conf[1])))
    print('Accuracy {0}'.format((conf[0,0] + conf[1,1])/sum(sum(conf))))
```

```python
# Calculate the models performance against the train set with a 
# threshold value of 0.45
conf_info(SongsLog4, 0.45)
```
