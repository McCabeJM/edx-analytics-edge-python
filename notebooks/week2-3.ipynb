```python
# Find which data columns are missing data values (NaN
# test_example = pd.isnull(train['fatherHS']).sum()
def nan_column_finder(df, columns):
    counts = {}
    for col in columns:
        nan_sum = pd.isnull(df[col]).sum()
        if nan_sum > 0:
            counts[col] = nan_sum
    return counts
        
```

```python
# Get the column variable names
cols = list(train.columns)
cols
```

```python
# Find which columns are missing values
nan_cols = nan_column_finder(train, cols)
nan_cols
```

```python
# Try to remove NaN values from columns
# Training set
train_without_nan = train.dropna()
print "Training set now has %s observations." % (train_without_nan.shape[0])
test_without_nan = test.dropna()
print "Testing set now has %s observations." % (test_without_nan.shape[0])
```

```python
# See if a pandas regression will automatically take care of reference levels
# Make a model for predicting student reading score
ind_vars = ' + '.join(cols[:len(cols) - 1])
score_formula = 'readingScore ~ ' + ind_vars
lmScore = sm.ols(formula=score_formula, data = train_without_nan).fit()
lmScore.summary()
```

```python
# Find Sum of Squared Errors 
# and the Root Mean Sqaured Errors (RMSE)
SSE = lmScore.ssr
train_n = train_without_nan.shape[0]
RMSE = math.sqrt(SSE/train_n)
RMSE
```

```python
# Find the baseline reading score
baseline = train_without_nan['readingScore'].mean()
baseline
```

```python
# Find SSE value for model
sse = ((test_without_nan.readingScore - lmScore.predict(test_without_nan))**2).sum()
sse
```

```python
# Find the SST value (Sum of Squared Errors for baseline model)
sst = ((test_without_nan.readingScore - train_without_nan.readingScore.mean())**2).sum()
sst
```

```python
# Find R-squared value
RSQ = 1 - (sse/sst)
RSQ
```
