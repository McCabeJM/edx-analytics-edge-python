```python
# Analytics Edge course from EdX
# Week 2 - Part 2
# Analysis of Climate Change Data from Climatic Research Unit 
# at the University of East Anglia

from pandas import DataFrame, read_csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as sm
%matplotlib inline
```

```python
# csv file location
climate_data_path = r'../csvs/climate_change.csv'
```

```python
# Read data into a data frame
climate = read_csv(climate_data_path)
# Check the data frame
climate.head()
```

```python
# Rename CFC-11 and CFC-12 column names for ease of use
climate = climate.rename(columns={'CFC-11': 'CFC11', 'CFC-12': 'CFC12'})
# Check
climate.head()
```

```python
# Build a training model with the data
# Climate data up to 2006
to_2006 = climate[climate['Year'] <= 2006]
# Check
to_2006.tail()
```

```python
# Make a subset for testing which includes climate data for the 
# Years from 2007 - 2008 (remaining years in the original data frame)
test_set = climate[climate['Year'] > 2006]
# Check
test_set.head()
```

```python
# Make the model
climate_model = sm.ols(formula='Temp ~ MEI + CO2 + CH4 + N2O + CFC11 + CFC12 + TSI + Aerosols', data=to_2006)
climate_model_fit = climate_model.fit()
climate_model_fit.summary()
```

```python
# Find the correlations between all variables in the data set
to_2006.corr()
```
