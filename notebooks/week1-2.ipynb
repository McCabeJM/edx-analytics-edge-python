```python
# Analytics Edge EdX course
# Part 2: Analyzes stock data from IBM, GE, P&G, Coca-Cola, and Boeing
# Datasets more specifically are from Infochimps at: 
# http://www.infochimps.com/datasets/nyse-daily-1970-2010-open-close-high-low-and-volume
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
ibm_file = 'IBMStock.csv'
ge_file = 'GEStock.csv'
pg_file = 'ProcterGambleStock.csv'
coke_file = 'CocaColaStock.csv'
boeing_file = 'BoeingStock.csv'

# Dataframes
IBM = read_csv(ibm_file)
GE = read_csv(ge_file)
ProctorGamble = read_csv(pg_file)
CocaCola = read_csv(coke_file)
Boeing = read_csv(boeing_file)

# Check the dataframe dimensions
print IBM.shape, GE.shape, ProctorGamble.shape, CocaCola.shape, Boeing.shape
```

```python
# Put dataframes into a list
stock_dataframes = [IBM, GE, ProctorGamble, CocaCola, Boeing]
```

```python
# Function to split the date variable in each dataframe into month, day, and year variables.
# Add these to each dataframe
def date_converter(dataframe):
    datetimes = pd.to_datetime(dataframe['Date'])
    dataframe['Month'] = [datetime.month for datetime in datetimes]
    dataframe['Day'] = [datetime.weekday() for datetime in datetimes]
    dataframe['Year'] = [datetime.year for datetime in datetimes]
    dataframe['Datetimes'] = datetimes
```

```python
# Function to convert dates for all dataframes
def complete_date_converter(dataframes):
    for dataframe in dataframes:
        date_converter(dataframe)
```

```python
# Convert date variable in all 5 company stock dataframes
complete_date_converter(stock_dataframes)
```

```python
# Make a function to create a dataframe from a certain range of dates
def date_range_dataframe(dataframe, begin, end):
    ranged_dataframe = dataframe[dataframe['Year'].isin(range(begin, end))]
    return ranged_dataframe
```

```python
# Assign date range data from 1995-2005 for each stock dataframe to a new 'ranged' dataframe
IBM_1995_2005, GE_1995_2005, PG_1995_2005, COKE_1995_2005, BO_1995_2005 = [date_range_dataframe(df, 1995, 2006) for df in stock_dataframes]
stock_data_1995_2005 = [IBM_1995_2005, GE_1995_2005, PG_1995_2005, COKE_1995_2005, BO_1995_2005]
```

```python
# Make function to find montly averages for each companies stock price
def monthly_mean_stockprice(company_df):
    montly_mean = company_df.groupby(['Month'])['StockPrice'].mean()
    return montly_mean
```

```python
# Function to find monthly mean stock prices for all companies and combine them to one list
ibm_monthly_mean_sp, ge_monthly_mean_sp, pg_monthly_mean_sp, coke_monthly_mean_sp, bo_monthly_mean_sp = [monthly_mean_stockprice(df) for df in stock_dataframes]  
```
